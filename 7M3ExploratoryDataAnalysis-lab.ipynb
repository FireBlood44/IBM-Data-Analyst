## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(x=df['ConvertedComp'], color='yellow')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)

IQR = Q3 - Q1
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print('the lower bound is ', lower, ' and the upper is ', upper)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = df[(df['ConvertedComp'] < lower) | (df['ConvertedComp'] > upper)]
print('total outliers = ', len(outliers))
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
new_df = df[(df['ConvertedComp'] >= lower) & (df['ConvertedComp'] <= upper)]
new_df['ConvertedComp'].describe()
```
